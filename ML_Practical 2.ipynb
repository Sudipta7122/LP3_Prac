### Spiliting data into input and output

```python
x = df.drop(['Email No.','Prediction'],axis = 1)
x
```

```python
y = df['Prediction']
y
```

### Checking the balancing of data

```python
sns.countplot(x=y)
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
```

```python
x_scale = scaler.fit_transform(x)
x_scale
```

### Cross validation

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(x_scale,y, random_state=7,test_size=0.07)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Initialize and build the model

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier()
```

```python
knn.fit(X_train,Y_train)
```

```python
Y_train=knn.predict(X_test)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay,accuracy_score,classification_report
```

```python
ConfusionMatrixDisplay.from_predictions(Y_test,y_pred)
```

```python
accuracy_score(Y_test,y_pred)
```

```python
print(classification_report(Y_test,Y_train))
```
